```python
import numpy as np
from collections import Counter

def predict_knn(X_train, y_train, query_point, k):
    """
    Predicts the class label for a single query point using the KNN algorithm.

    Args:
        X_train (pd.DataFrame or np.ndarray): Training features.
        y_train (pd.Series or np.ndarray): Training target variable.
        query_point (pd.Series or np.ndarray): The point for which to predict the class.
        k (int): The number of nearest neighbors to consider.

    Returns:
        The predicted class label for the query point.
    """
    # Calculate Euclidean distances between the query point and all training points
    distances = np.sqrt(np.sum((X_train - query_point)**2, axis=1))

    # Combine distances with class labels
    distances_with_labels = list(zip(distances, y_train))

    # Sort by distance
    sorted_distances_with_labels = sorted(distances_with_labels)

    # Get the class labels of the k nearest neighbors
    k_nearest_neighbors = sorted_distances_with_labels[:k]
    k_nearest_labels = [label for distance, label in k_nearest_neighbors]

    # Determine the most frequent class among the k neighbors
    most_common_class = Counter(k_nearest_labels).most_common(1)[0][0]

    return most_common_class

# Example usage (you can replace this with your own query point)
# Let's use the first point from the test set as an example query point
query_example = X_test.iloc[0]
predicted_class = predict_knn(X_train, y_train, query_example, k=5)

display(f"Example Query Point: {query_example.values}")
display(f"Predicted Class (using custom KNN function): {predicted_class}")
display(f"Actual Class (from y_test): {y_test.iloc[0]}")
```
