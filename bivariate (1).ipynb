```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = sns.load_dataset('titanic')
df.head()
```

```python
# Creating a custom Car Prices dataset
data = {
    'Car_Model': ['Honda City', 'Hyundai i20', 'Maruti Swift', 'Toyota Fortuner',
                  'BMW X1', 'Audi A4', 'Tata Nexon', 'Kia Seltos', 'Mahindra XUV700', 'Renault Kwid'],
    'Engine_Size_CC': [1498, 1197, 1198, 2755, 1995, 1984, 1497, 1493, 2184, 999],
    'Mileage_kmpl': [17.8, 20.0, 21.2, 12.0, 16.0, 15.5, 17.0, 18.4, 14.0, 22.3],
    'Horsepower': [121, 83, 89, 201, 187, 188, 118, 115, 200, 68],
    'Price_Lakhs': [14.5, 9.0, 8.5, 45.0, 50.0, 47.0, 13.5, 16.0, 27.0, 5.5]
}

df = pd.DataFrame(data)

# Display dataset
print("Car Prices Dataset:")
df
```

```python
# Display basic info about the dataset
print("Dataset Info:")
print(df.info())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Summary statistics for numerical columns
print("Summary Statistics:")
df.describe()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Engine_Size_CC', y='Price_Lakhs', hue='Mileage_kmpl', palette='coolwarm', s=100)
plt.title('Engine Size vs Price (colored by Mileage)')
plt.xlabel('Engine Size (CC)')
plt.ylabel('Price (Lakhs)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Horsepower', y='Price_Lakhs', hue='Car_Model', s=100)
plt.title('Horsepower vs Price by Car Model')
plt.xlabel('Horsepower (HP)')
plt.ylabel('Price (Lakhs)')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
sns.pairplot(df[['Engine_Size_CC', 'Mileage_kmpl', 'Horsepower', 'Price_Lakhs']])
plt.suptitle('Pairwise Relationships Between Car Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['Engine_Size_CC', 'Mileage_kmpl', 'Horsepower', 'Price_Lakhs']].corr(),
            annot=True, cmap='Blues', linewidths=0.5)
plt.title('Correlation Heatmap for Car Features')
plt.show()
```

```python
print("""
Observations:
1. Price is strongly correlated with Engine Size and Horsepower  cars with bigger engines and more power cost more.
2. Mileage shows a weak negative correlation with Price  fuel-efficient cars tend to be cheaper.
3. Luxury models (BMW, Audi, Fortuner) have higher horsepower, larger engines, and higher prices.
4. Compact cars (Swift, Kwid, i20) have smaller engines, higher mileage, and lower prices.
""")
```
